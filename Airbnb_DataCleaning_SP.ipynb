```python
from IPython.display import display
import nltk
import inspect
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cmx
import matplotlib.colors as colors
import pandas as pd
from collections import Counter
%matplotlib inline
```

```python
#get rid of NaN values that can't be salvaged
df = pd.read_csv('listings.csv',encoding='latin1')
df = df[(pd.notnull(df['id']))&(pd.notnull(df['host_id']))&(pd.notnull(df['zipcode']))&(pd.notnull(df['latitude']))]
df = df[(pd.notnull(df['longitude']))&(pd.notnull(df['bathrooms']))&(pd.notnull(df['bedrooms']))&(pd.notnull(df['beds']))]
df = df[(pd.notnull(df['property_type']))&(pd.notnull(df['price']))&((df.number_of_reviews!=0)&(pd.notnull(df.review_scores_rating)))]
```

```python
#handle data type problems

zipcodes = []
weekly_prices = []
monthly_prices =[]
extra_p = []
prices = []
for index,row in df.iterrows():
    zipc = str(row.zipcode)[:5]
    price = int(str(row.price).replace('$','').replace(',','').partition('.')[0])
    extra = int(str(row.extra_people).replace('$','').replace(',','').partition('.')[0])
    if pd.isnull(row.weekly_price):
        wprice = price * 7
    else:
        wprice = int(str(row.weekly_price).replace('$','').replace(',','').partition('.')[0])
    if pd.isnull(row.monthly_price):
        mprice = price * 30
    else:
        mprice = int(str(row.monthly_price).replace('$','').replace(',','').partition('.')[0])
    weekly_prices.append(wprice)
    monthly_prices.append(mprice)
    prices.append(price)
    extra_p.append(extra)
    zipcodes.append(zipc)

df['zipcode'] = pd.Series(zipcodes)
df['weekly_price'] = pd.Series(weekly_prices)
df['monthly_price'] = pd.Series(monthly_prices)
df['price'] = pd.Series(prices)
df['extra_people'] = pd.Series(extra_p)
```
